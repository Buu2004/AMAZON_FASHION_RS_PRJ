```python
import numpy as np
import pandas as pd
import bar_chart_race as bcr
```

```python
path = 'merge_5.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df['reviewTime'] = pd.to_datetime(df['reviewTime'], format='%m %d, %Y')
df['year_month'] = df['reviewTime'].dt.to_period('M')
df
```

```python
df = df.sort_values('year_month')
df.drop('title',axis='columns', inplace=True)
df.drop('reviewTime',axis='columns',inplace=True)
df.drop('Unnamed: 0',axis='columns',inplace=True)
df.drop('unixReviewTime',axis='columns',inplace=True)
df
```

```python
yearMonth_uni = np.array(df['year_month'].unique().tolist())
yearMonth_uni.shape
```

```python
asin_uni = np.array(df['asin'].unique().tolist())
asin_uni.shape
```

```python
df_new = pd.DataFrame()
for asin in asin_uni:
    df_cc = df.loc[df['asin'] == asin]
    yearMonth = np.array(df_cc['year_month'].unique().tolist())
    overall_sum = 0
    count = 0
    lis = []
    mean = 0
    n = 0
    for i in yearMonth_uni:
        if df_cc.loc[df_cc['year_month'] == i ]['overall'].count() == 0:
            lis.append(overall_sum)
        else:
            count += df_cc.loc[df_cc['year_month'] == i ]['overall'].count()
            overall_sum += df_cc.loc[df_cc['year_month'] == i ]['overall'].sum()
            # mean = overall_sum/count
            lis.append(overall_sum)
    print(lis)
    df_new[asin] = lis


df_new.index = yearMonth_uni
df_new
```
