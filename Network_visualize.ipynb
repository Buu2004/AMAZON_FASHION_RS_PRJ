```python
import pandas as pd

from pyvis.network import Network
```

```python
file_path = "D:\\1 Ép bê tê\\4 Kỳ 4 DE\\DAP\\Data\\Data_project\\AMAZON_FASHION_5.csv"
data_fashion = pd.read_csv(file_path)
```

```python
data_fashion = pd.DataFrame(data_fashion)
data_fashion.head()
```

```python
df = pd.DataFrame()

df['ID customer'] = data_fashion['reviewerID']
df['Customer Name'] = data_fashion['reviewerName']
df['Product ID'] = data_fashion['asin']
df
```

```python
customer_relations = []
size = 1
for i in range(df.shape[0]):
    # Lấy thông tin của khách hàng hiện tại
    current_customer_id = df.at[i, 'ID customer']
    current_customer_name = df.at[i, 'Customer Name']
    current_product_id = df.at[i, 'Product ID']

    # Duyệt qua các dòng còn lại để tìm khách hàng có sản phẩm giống nhau
    for j in range(i + 1, df.shape[0]):
        if df.at[j, 'Product ID'] == current_product_id:

            # Lấy thông tin của khách hàng có sản phẩm giống nhau
            related_customer_id = df.at[j, 'ID customer']
            related_customer_name = df.at[j, 'Customer Name']

            # Tìm index của quan hệ trong danh sách
            index_to_replace = None
            for idx, relation in enumerate(customer_relations):
                if relation['source'] == current_customer_name and relation['target'] == related_customer_name:
                    index_to_replace = idx
                    break

            # Nếu quan hệ đã tồn tại, thay thế giá trị size
            if index_to_replace is not None:
                customer_relations[index_to_replace]['size'] += size
            else:
                # Nếu quan hệ chưa tồn tại, thêm mới vào danh sách
                customer_relations.append({
                    'source': current_customer_name,
                    'target': related_customer_name,
                    'size': size  # Trọng số
                })

df_relations = pd.DataFrame(customer_relations)
```

```python
sample = df_relations.sample(1000, random_state = 0)
sample
```

```python
net = Network(
    notebook = True,
    cdn_resources = 'remote',
    bgcolor = "#222222",
    font_color = "white",
    height = "750px",
    width = "100%",
    select_menu = True,
    filter_menu = True
)
```

```python
node = list(set([*sample.source, *sample.target]))
edge = sample.values.tolist()
net.add_nodes(node)
net.add_edges(edge)
```

```python
net.options.layout = {
    'hierarchical': {
        'enabled': True,
        'direction': 'LR',
        'sortMethod': 'directed',
        'nodeSpacing': 150
    }
}

net.show('network.html')
```
